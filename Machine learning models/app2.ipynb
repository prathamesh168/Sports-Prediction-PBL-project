### spliting the data into training and testing 

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,stratify=y, random_state=42)
```

### Importing Tensorflow squential model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow import keras
from tensorflow.keras.layers import BatchNormalization
```

### creating the sequential model with 2 hidden layers


```python
model = Sequential()

model.add(Dense(units=50,kernel_initializer="he_normal",input_shape=X_train.shape[1:],kernel_regularizer=keras.regularizers.l2(0.001)))
model.add(BatchNormalization())
model.add(keras.layers.Activation('relu'))
model.add(Dropout(0.2))

model.add(Dense(units=50,kernel_initializer="he_normal",kernel_regularizer=keras.regularizers.l2(0.001)))
model.add(BatchNormalization())
model.add(keras.layers.Activation('relu'))
model.add(Dropout(0.2))

# model.add(Dense(units=50,kernel_initializer="he_normal",kernel_regularizer=keras.regularizers.l2(0.001)))
# model.add(BatchNormalization())
# model.add(keras.layers.Activation('relu'))
# model.add(Dropout(0.2))

model.add(Dense(units=8, activation="softmax",kernel_initializer="GlorotNormal",kernel_regularizer=keras.regularizers.l2(0.001)))
```

### created model summery 

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
opt = Adam(learning_rate=0.005)
model.compile(optimizer=opt, loss="categorical_crossentropy", metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
```

### using early stopping avoiding overfitting 

```python
early_stopping_cb = keras.callbacks.EarlyStopping(patience=20,
 restore_best_weights=True)
```

### training the model for training data and validating over test data

```python
predict = model.fit(X_train, y_train, epochs=200,
 validation_data=(X_test, y_test),
 callbacks=[early_stopping_cb])
```
